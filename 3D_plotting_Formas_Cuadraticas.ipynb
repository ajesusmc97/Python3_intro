## Parte I. Superficie de una forma cuadr치tica 

En este ejercicio se graficar치 la forma cuadr치tica:

$$ Q(x,y) = ax^2 + bxy + cy^2 $$ 
con las herramientas de Matplotlib 

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
import matplotlib 
matplotlib.rcParams['text.usetex'] = True  #Tipograf칤a Serif de LATeX 
matplotlib.rc('xtick', labelsize=12) 
matplotlib.rc('ytick', labelsize=12)   #Tama침o de los ejes 

fig = plt.figure(figsize=(15,10))
ax = fig.gca(projection='3d')

#Datos
a= 10
b= 1
c= 10
X =np.arange(-10, 10, 0.25)
Y = np.arange(-10, 10, 0.25)
X, Y = np.meshgrid(X, Y)
Z = a*X**2 + b*X*Y + c*Y**2


#Plotear

superficie = ax.plot_surface(X,Y,Z, cmap=cm.coolwarm, linewidth=1 , cstride=1 , rstride = 1)
plt.title("Gr치fica de una forma cuadr치tica",size=20)
plt.xlabel("Valores de x",size=20)
plt.ylabel("Valores de y",size=20 )
#fig.colorbar(superficie,shrink =0.5,aspect=5)  
plt.show() 
```

# Parte II . Mapas de contorno 

Ahora ser치n graficados los mapas de contorno de la forma cuadr치tica 


```python
fig,ax=plt.subplots(1,1 , figsize=(10,8))
cp = ax.contourf(X, Y, Z, cmap=cm.coolwarm)
# fig.colorbar(cp) # Barra de color 
#ax.clabel(cp)
ax.set_title('Mapas de contorno de la forma cuadr치tica',size=14)
ax.set_xlabel('x (cm)', size=20)
ax.set_ylabel('y (cm)', size=20)
plt.show()
```
