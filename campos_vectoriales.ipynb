# Campos vectoriales en Python 

## Parte I. Gráfico de un campo vectorial con quiver

El campo magnético que genera un alambre recto viene dado por las ecuaciones: 

$$ \vec{B} = (B_x,B_y) $$ 

Donde:

$$ B_x = -\cfrac{y}{x^2 + y^2} $$ 
$$ B_y = \cfrac{x}{x^2 + y^2} $$ 


```python
import matplotlib.pyplot as plt   #Solo necesitamos estas librerías 
import numpy as np 


```

```python
x = np.linspace(-10,10,15) 
y = np.linspace(-10,10,15)
X,Y = np.meshgrid(x,y)
r2 = X**2 + Y**2 
```

```python
Bx = -Y/(r2 + 0.0001)
By = X/(r2+0.0001)
plt.figure(figsize=(10,8))  
plt.axis([-14,14,-14,14])  #Longitudes de los ejes 
plt.title("Gráfica del campo vectorial magnético de un alambre",size=15)
QP = plt.quiver(X,Y,Bx,By,color="blue") 
plt.quiverkey(QP,0.9,0.9,1.0,"campo magnetico")
```

## Parte II. Conjuntos de nivel y campos vectoriales. 

Supongamos que tenemos la función campo escalar: 
$$ f(x,y) = \cfrac{1}{x^2 + y^2 + 1} $$ 

De la que deseamos graficar sus conjuntos de nivel y su campo vectorial gradiente.

Primero graficaremos la función utilizando matplitlib  

```python

from mpl_toolkits.mplot3d import Axes3D   #Librería que permite graficar en 3D 
fig = plt.figure(figsize=(10,8))
ax=fig.gca(projection="3d")


#Dominio de la función 
A= np.linspace(-10,10,200)
B = np.linspace(-10,10,200) 
X,Y = np.meshgrid(A,B)   # Dominio de la función 
Z = 1/(X**2 + Y**2 + 1 )  
plt.title("Gráfica del campo escalar")
grafica = ax.plot_surface(X,Y,Z, cstride=1,rstride=1, linewidth=10)
```

```python
fig,ax=plt.subplots(1,1 , figsize=(10,8))
A = ax.contourf(X,Y,Z)
```
